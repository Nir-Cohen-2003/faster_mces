```python
# analyze the clusters- find the ocnnected components in the mces matrix, and print the histogram of the sizes of the connected components
# Set the threshold for determining connected components
threshold = 5  # You can adjust this value as needed
def analyze_clusters(mces_matrix, threshold):
# Find pairs with MCES lower bound less than the threshold
    not_distinct = (mces_matrix < threshold)
    rows, cols = np.where(not_distinct)
    edge_count = len(rows)

    # Create a sparse adjacency matrix for the graph
    n = mces_matrix.shape[0]
    not_distinct_sparse = sp.csr_matrix((np.ones(edge_count, dtype=np.int8), (rows, cols)), shape=(n, n))

    # Find connected components
    _, labels = connected_components(csgraph=not_distinct_sparse, directed=False, return_labels=True)

    # Compute sizes of each connected component
    component_sizes = np.bincount(labels)

    # Print histogram of the sizes of the connected components
    print(f"Connected component sizes for threshold {threshold}:")
    print(list(sorted(component_sizes, reverse=True)))
    return component_sizes

for threshold in range(4, 11):
    analyze_clusters(mces_matrix, threshold)
```
